```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load your cleaned data
df = pd.read_csv('/users/shivamkumarsingh/desktop/amazon_sales.csv')

# Step 1: Convert 'order_date' to datetime and extract useful numeric features
df['order_date'] = pd.to_datetime(df['order_date'])
df['order_date_ts'] = df['order_date'].map(pd.Timestamp.timestamp)
df['order_day'] = df['order_date'].dt.day
df['order_weekday'] = df['order_date'].dt.weekday

# Optionally drop original 'order_date' column
df = df.drop('order_date', axis=1)

# Step 2: One-hot encode categorical variables
df = pd.get_dummies(df, columns=['city', 'category', 'item', 'month'], drop_first=True)

# Step 3: Define features (X) and target (y)
X = df.drop('quantity', axis=1)  # Predicting 'quantity' here as an example
y = df['quantity']

# Step 4: Split data into train and test sets (70%-30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 5: Train the Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Step 6: Make predictions
y_pred = model.predict(X_test)

# Step 7: Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print('Mean Squared Error:', mse)
print('R2 Score:', r2)
```

```python
plt.scatter(y_test, y_pred, color='blue')
plt.xlabel('Actual Sales')
plt.ylabel('Predicted Sales')
plt.title('Actual vs Predicted Sales')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')  # Diagonal line
plt.show()
```

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('/users/shivamkumarsingh/desktop/amazon_sales.csv')

# Convert 'order_date' to datetime and set as index
df['order_date'] = pd.to_datetime(df['order_date'])
ts_data = df.set_index('order_date')['sales'].sort_index()

# Train-test split chronologically (70%-30%)
train_size = int(len(ts_data) * 0.7)
train, test = ts_data[:train_size], ts_data[train_size:]

# Fit ARIMA model - example order (p,d,q) = (2,1,2)
model = ARIMA(train, order=(2,1,2))
model_fit = model.fit()

# Forecast test period
forecast = model_fit.forecast(steps=len(test))

# Evaluate
mse = mean_squared_error(test, forecast)
print('Test MSE:', mse)

# Plot actual vs forecast
plt.figure(figsize=(10,6))
plt.plot(test.index, test, label='Actual Sales')
plt.plot(test.index, forecast, label='Forecasted Sales', linestyle='--')
plt.title('ARIMA Model Forecast vs Actual Sales')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
import pandas as pd

# Load the dataset (use your CSV file path)
df = pd.read_csv('/users/shivamkumarsingh/desktop/amazon_sales.csv')

# Map month names to numbers
month_map = {'January':1, 'February':2, 'March':3, 'April':4, 'May':5, 'June':6,
             'July':7, 'August':8, 'September':9, 'October':10, 'November':11, 'December':12}

df['month_num'] = df['month'].map(month_map)

# Create datetime column at month start
df['date'] = pd.to_datetime({'year': df['year'], 'month': df['month_num'], 'day': 1})

# Aggregate monthly sales time series with monthly frequency
monthly_sales = df.groupby('date')['sales'].sum().asfreq('MS').ffill()

print(monthly_sales.head())
```

```python
from statsmodels.tsa.stattools import adfuller

result = adfuller(monthly_sales)
print(f'ADF Statistic: {result[0]}')
print(f'p-value: {result[1]}')

if result[1] < 0.05:
    print("The series is stationary.")
else:
    print("The series is non-stationary - differencing is likely needed.")
```

```python
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plot_acf(monthly_sales)
plt.show()

plot_pacf(monthly_sales)
plt.show()
```

```python
diff_series = monthly_sales.diff().dropna()

result_diff = adfuller(diff_series)
print(f'ADF Statistic (1st difference): {result_diff[0]}')
print(f'p-value (1st difference): {result_diff[1]}')

if result_diff[1] < 0.05:
    print("Differenced series is stationary.")
else:
    print("Differenced series remains non-stationary; consider higher differencing.")
```

```python
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plot_acf(diff_series)
plt.title('ACF of Differenced Series')
plt.show()

plot_pacf(diff_series)
plt.title('PACF of Differenced Series')
plt.show()
```

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Assume monthly_sales is your stationary time series with monthly frequency

# Train-test split (e.g., 70% train)
train_size = int(len(monthly_sales) * 0.7)
train, test = monthly_sales[:train_size], monthly_sales[train_size:]

# You determined d=1 from differencing test
# Select p and q based on your ACF and PACF analysis; example p=1, q=1
p, d, q = 1, 1, 1

# Train ARIMA model
model = ARIMA(train, order=(p, d, q))
model_fit = model.fit()

# Forecast
forecast = model_fit.forecast(steps=len(test))

# Evaluate
mse = mean_squared_error(test, forecast)
r2 = r2_score(test, forecast)
print(f'Test MSE: {mse}')
print(f'Test R2: {r2}')

# Plot actual vs forecast
plt.figure(figsize=(10, 6))
plt.plot(train.index, train, label='Train')
plt.plot(test.index, test, label='Test')
plt.plot(test.index, forecast, label='Forecast', linestyle='--')
plt.legend()
plt.show()
```

```python
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Assume monthly_sales is your time series with monthly frequency

# Train-test split
train_size = int(len(monthly_sales) * 0.7)
train, test = monthly_sales[:train_size], monthly_sales[train_size:]

# SARIMA order parameters:
# (p,d,q) for non-seasonal and (P,D,Q,s) for seasonal part
# Example: seasonal period s=12 for monthly data with yearly seasonality
p, d, q = 1, 1, 1
P, D, Q, s = 1, 1, 1, 12

# Fit SARIMA model
model = SARIMAX(train, order=(p,d,q), seasonal_order=(P,D,Q,s))
model_fit = model.fit()

# Forecast test period
forecast = model_fit.forecast(steps=len(test))

# Evaluate
mse = mean_squared_error(test, forecast)
r2 = r2_score(test, forecast)
print(f'SARIMA Test MSE: {mse}')
print(f'SARIMA Test R2: {r2}')

# Plot
plt.figure(figsize=(10,6))
plt.plot(train.index, train, label='Train')
plt.plot(test.index, test, label='Test')
plt.plot(test.index, forecast, label='Forecast', linestyle='--')
plt.title('SARIMA Forecast vs Actual')
plt.legend()
plt.show()
```
